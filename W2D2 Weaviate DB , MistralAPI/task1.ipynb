```python
# Define the schema for the class 'File'
class_obj={
    "class": "File",
    "properties": [
        {
            "name": "title",
            "dataType": ["string"]
        },
        {
            "name": "description",
            "dataType": ["text"]
        }
    ],
    # Configure to use Mistral embeddings
    "vectorizer": "text2vec-mistral",
    "vectorIndexConfig": {
        "vectorizer": {
            "property": "description"  # Embedding will be generated from 'description'
        }
    }
}
client.schema.create_class(class_obj)
```

```python
# Load the PDF using PyMuPDFLoader
loader = PyMuPDFLoader("C:/Users/muaaz/Uni_info.pdf")
docs = loader.load()

# Set up chunking configuration
chunk_size = 500
chunk_overlap = 50

# Initialize the text splitter
splitter = CharacterTextSplitter(
    separator='\n',
    chunk_size=chunk_size,
    chunk_overlap=chunk_overlap,
    length_function=len
)

# Start the batch process with the specified batch size
batch = client.batch.configure(batch_size=100)  # Initialize the batch object

with client.batch as batch:
    for page in docs:
        page_splits = splitter.split_text(page.page_content)
        for split in page_splits:
            # Prepare the data object with title and description
            proporties = {
                "title": "uniinfo",  # Title for each object
                "description": split  # Chunked text for embedding
            }
            # Add the object to the batch (no references provided)
            batch.add_data_object(data_object=proporties, class_name="File" )
```
